```python
import ytil.matplotlib as visual
from PIL import Image
import numpy as np
import time
from os.path import join
from matplotlib import pyplot
src_dir = '/media/sec/landcover/large/val/src/'
lbl_dir = '/media/sec/landcover/large/val/lbl/'
dlabv3_pred_dir = '/home/yhong/work/landcover/src/detect/lulc/dlabv3_pred/'
dan_pred_dir = '/home/yhong/work/landcover/src/detect/lulc/dan_pred/'
fdn_pred_dir = '/home/yhong/work/landcover/src/detect/lulc/fdn_pred/'
from IPython.display import display, clear_output
def img_name(idx):
    return str(idx)+'.png'
def get_image(data_dir, idx):
    path = join(data_dir, img_name(idx))
    return np.array( Image.open(path) )
def combine(images):
    return np.concatenate(images, axis=0)
        
#ax.cla()
#https://blog.shahinrostami.com/2018/09/jupyter-notebook-and-updating-plots/

idx = 1999
#idx = 3699
for idx in range(1000):
    time.sleep(1)
    src = get_image(src_dir, idx)
    gt = get_image(lbl_dir, idx)
    dlab = get_image(dlabv3_pred_dir, idx)
    dan = get_image(dan_pred_dir, idx)
    fdn = get_image(fdn_pred_dir, idx)
    labels = combine([gt, dlab, fdn, dan])
    #print(labels.shape)
    visual.pyplot_show_images( [src, labels] )
    clear_output(wait=True)
    pyplot.show()
    
```
